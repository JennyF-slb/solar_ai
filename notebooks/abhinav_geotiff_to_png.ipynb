```python
import os
from PIL import Image

tif_folder_path = '/content/Colab Notebooks/project-solar-ai-solutions/AerialImageDataset/train/images'  # Path to the folder containing TIF images (Geo-tif)
png_output_folder_path = '/content/Colab Notebooks/project-solar-ai-solutions/AerialImageDataset_png/train/png_images'  # Path to the folder for saving PNG images

# Create the output folder if it doesn't exist
os.makedirs(png_output_folder_path, exist_ok=True)

# Convert TIF files to PNG and save in the output folder
for filename in os.listdir(tif_folder_path):
    if filename.endswith(".tif"):
        tif_path = os.path.join(tif_folder_path, filename)
        png_filename = filename[:-4] + ".png"
        png_path = os.path.join(png_output_folder_path, png_filename)
        img = Image.open(tif_path)
        img.save(png_path, "PNG")
```
